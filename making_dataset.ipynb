```python
from pandas import read_csv
df = read_csv('main_dataset.csv')
```

```python
df
```

```python
# data = df['title'] + ', ' + df['text']  # + ', ' + df['Tags']
# target = df['tags']
data = df['Title'] + ', ' + df['Abstract']  # + ', ' + df['Tags']
target = df['Class']
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.3, shuffle=True, random_state=0, stratify=target)
```

```python
import utils1 as prof
estimator = prof.ProfileClassifierDefault(profile_type=prof.ProfileType.MI, profile_length=400, max_vocabulary_size=5000)
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.set_profile_type(prof.ProfileType.S_JOINT_OCCUR)
estimator.set_profile_length(7)
```

```python
for cls, profile in estimator.profiles.items():
    print(f"{cls}: weights size = {len(profile.all_words)}\n{profile.weights}\n\n")
```

```python
estimator.set_profile_type(prof.ProfileType.SS2_SOKAL_SNIS_2)
estimator.set_profile_length(7)
for cls, profile in estimator.profiles.items():
    print(f"{cls}: weights size = {len(profile.all_words)}\n{profile.weights}\n\n")
```

```python
from time import time
t1 = time()
prediction = estimator.predict(X_test)
t2 = time()
print(f"prediction time = {t2 - t1}")
for probabilities, predicted_cls, true_cls in zip(estimator.predict_proba(X_test), prediction, y_test):
    print(f"{probabilities}: \n\t\tprediction = {predicted_cls}\n\t\ttarget = {true_cls}\n\n")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, prediction))
```

```python
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt

for profile_type in [prof.ProfileType.S_JOINT_OCCUR]:  #
    estimator.set_profile_type(profile_type)
    f1_scores = list()
    lengths = list()
    best_f1 = 0
    best_length = 0
    for length in range(1, 16):
        estimator.set_profile_length(length)
        prediction = estimator.predict(X_test)
        f1 = f1_score(y_test, prediction, average='weighted')
        if f1 > best_f1:
            best_length = length
            best_f1 = f1
        f1_scores.append(f1)
        lengths.append(length)
        print(f"type = {profile_type.name}, length = {length}, score = {f1}")
    plt.plot(lengths, f1_scores, label=profile_type.name)
plt.xlabel('Длина профиля')
plt.ylabel('f1-мера')
plt.legend()
plt.show()
```

```python
print(f"best: {estimator.profile_type}; score={best_f1}; length={best_length}")
```
